```python
import pandas as pd
```

```python
df = pd.read_excel(r"C:\Users\ASUS\Downloads\data_cleaning .xlsx", sheet_name="Usage",header = None, names = ['Raw'])
df
```

```python
#deleting spaces and _ on both sides of string
df['Raw'] = df['Raw'].str.strip('_ ')
```

```python
#splitting into two columns date and energy
df[['Datetime','Energy']] = df['Raw'].str.split(r'_+', n=1, expand = True) 
```

```python
#extracting usage kwh
df["Usage_Kwh"] = df['Energy'].str.extract(r'([\d\.]+)').astype(float)
df.drop(["Energy"], axis = 1, inplace = True)
```

```python
#Extracting only hours in "XPM/XAM"
df["Hour"] = df['Datetime'].str.extract(r'(\d{1,2}\s*([a-zA-Z]{2}))')[0].str.replace(" ","")
```

```python
df
```

```python
#Extracting date
df['Datetime'] = df['Datetime'].str.extract(r'(\d{1,2}[a-zA-Z]{0,2}-[a-zA-Z]{3}-\d{4})')
```

```python
#Deleting unnecessary date characters
df['Datetime'] = df['Datetime'].str.replace(r'(st|nd|rd|th)', '', regex=True)
```

```python
#Converting to datetime
df['Datetime'] = pd.to_datetime(df['Datetime'] + " " + df['Hour'], format='%d-%b-%Y %I%p')
```

```python
#Creating new date columns in order to better analyze
df['Month'] = df['Datetime'].dt.month_name()
df['Day_of_Week'] = df['Datetime'].dt.day_name()
df['Day'] = df['Datetime'].dt.day
df['Hour'] = df['Datetime'].dt.hour
```

```python
df
```

```python
#Dropping unnecessary column
df.drop('Raw', axis = 1, inplace = True)
```

```python
#Sorting by datetime
df = df.sort_values('Datetime').reset_index(drop = True)
```

```python
#The data is ready for analysis and answering questions
df
```

```python
#Q1
#What is your average hourly electricity usage?
print(df['Usage_Kwh'].mean().round(3),"kwh") # The answer is B
```

```python
#Q2
#What is your average electricity usage perhour in February?
df[df['Month'] == 'February']['Usage_Kwh'].mean().round(3) # The answer is D
```

```python
#Q3
#Which day of the week has the highest average usage
df.groupby('Day_of_Week')['Usage_Kwh'].mean().idxmax() # The answer is A
```

```python
#Q4
#What is the highest amount of electricity used in a continuous 4 hour period?
df['Usage_Kwh'].rolling(window=4).sum().max() # The answer is A
```

```python
#Question 5 Based on your historic electricity usage, what would your annual cost of electricity be under the "Monthly Flex"?
df_contracts = pd.read_excel(r'c:\Users\ASUS\Downloads\data_cleaning .xlsx', sheet_name="Contracts")
```

```python
#Creating a dataframe for monthly flex
monthly_flex = df_contracts.iloc[11:23,4:6]
monthly_flex.columns =["Month","Rate"]
monthly_flex["Rate"] = monthly_flex["Rate"].astype(float)
monthly_flex
```

```python
#Creating a dataframe for usage by month and merging with monthly flex
usage_by_month = df.groupby('Month')['Usage_Kwh'].sum()
merged_data = pd.merge(usage_by_month, monthly_flex, on='Month')
merged_data
```

```python
#Calculating annual cost
merged_data['Cost'] = merged_data["Usage_Kwh"] * merged_data["Rate"]
monthly_cost = merged_data['Cost'].sum() 
print(monthly_cost) # The answer is B
```

```python
#Question 6 Based on your historic electricity usage, which of the three contracts would pro-duce the lowest annual cost?
# Monthly contract will be 1421.20577

#Flat rate
flat_rate_cost = df.groupby('Month')['Usage_Kwh'].sum().sum() * 0.21
print(flat_rate_cost)
#1438.10394 will be flat rate annual cost
```

```python
#Hourly contract
hourly_flex = df_contracts.iloc[11:35,8:10].reset_index(drop = True)
hourly_flex.columns = ["Hour","Rate"]
```

```python
#Creating a dataframe for hourly flex
hourly_flex['Hour'] = range(24) #This transforms our long string format hours to integer in order to merge with our main dataframe
hourly_flex['Hour'] = hourly_flex['Hour'].astype(int)
hourly_flex
```

```python
#Merging useful columns
hourly_cost = pd.merge(df[['Hour','Month',"Usage_Kwh"]], hourly_flex, on = 'Hour')
```

```python
#Calculating annual cost for hourly flex
hourly_cost['Cost'] = hourly_cost['Usage_Kwh'] * hourly_cost['Rate']
hourly_cost = hourly_cost['Cost'].sum()

#Comparison of contracts
print(hourly_cost) # The hourly cost has lowest annual cost relative to others then answer is C
print(monthly_cost)
print(flat_rate_cost)
```
